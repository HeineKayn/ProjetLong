```python
from mmap import mmap, PROT_READ
import re
import sys
import string

def strings(filename, min=5):
    with open(filename, errors="ignore") as f:
        result = ""
        for c in f.read():
            if c in string.printable:
                result += c
                continue
            if len(result) >= min:
                yield result
            result = ""
        if len(result) >= min:  # catch result at EOF
            yield result
```

```python
def isletter(x):
    cond = x.isdigit()
    y = ord(x)
    cond = cond or (y > 64 and y < 91)
    cond = cond or (y > 96 and y < 123)
    return cond
```

```python
path = "../data/samples/"
file = "pink.exe"

strings_dist={}
for word in strings(path+file):
    if word in strings_dist.keys():
        strings_dist[word] += 1
    else :
        strings_dist[word] = 1
print(strings_dist)
```
